```python
import pandas as pd
wine=pd.read_csv('wine.csv')
print(wine.head())



```

```python

x_quality=wine.loc[:,'quality']
print(x_quality)
x_sulphates=wine.iloc[:,0:11]
print(x_sulphates)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
xx= np.array([5,2,3,1])
yy=np.array([6,8,9,0])

mod = sm.OLS(x_quality,x_sulphates)
res = mod.fit()
res.summary()

fig,ax=plt.subplots(figsize=(20,6))


sm.graphics.plot_fit(res,8,ax=ax)
fig1,ax1=plt.subplots()


sm.graphics.plot_fit(res,9,ax=ax1)

plt.show()
```

```python
import matplotlib.pyplot as plt
yyy = [xxx for xxx in range(1599)]
plt.figure(figsize=(12,6))
plt.bar(yyy,wine['fixed acidity'])
plt.bar(wine['quality'],wine['fixed acidity'])

plt.show()
```

```python
from sklearn import linear_model
test=wine.loc[0:798,]
test_x=test.iloc[:,0:11]

train=wine.loc[800:1598,]
train_x=train.iloc[:,0:11]
model2 = linear_model.LinearRegression()
model2.fit( train_x, train['quality'])


plt.scatter(train['pH'], test['quality'],color='g')
plt.plot(train['pH'], model2.predict(test['quality']),color='k')

plt.show()
```
